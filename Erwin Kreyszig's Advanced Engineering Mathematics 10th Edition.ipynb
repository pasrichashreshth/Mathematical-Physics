```python
import numpy as np
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
def i(t):
    return t - t
def i_1(t):
    return 2.718**((1/2)-t)
def i_2(t):
    return -(2.718**((3/5)-t))
def v(t):
    return t/t

# Bounds of t
t = np.linspace(0, 1/2, 500)
t_1 = np.linspace(1/2, 3/5, 100)
t_2 = np.linspace(3/5, 3, 2400)

# Plot
plt.figure(figsize = (9, 9))
plt.plot(t, i(t), color = 'c', linewidth = 3)
plt.plot(t_1, i_1(t_1), color = 'c', linewidth = 3, label = 'Current')
plt.plot(t_2, i_1(t_2) + i_2(t_2), color = 'c', linewidth = 3)
plt.plot(t_1, v(t_1), color = 'b', linewidth = 3, label = 'Voltage')

# Attributes
plt.title('RC Circuit with Piecewise Constant Potential Input')
plt.xlabel('Time, t')
plt.ylabel('Current and Voltage')
plt.legend()

# Show the plot
plt.show()
```

## Problem 35 ##

```python
def i(t):
    return 100*np.sin(t) + 1000*np.sin(10*t)
def v(t):
    return -9900*np.cos(t)

# Bounds of t
t = np.linspace(np.pi, 3*(np.pi), 1000)

# Plot
plt.figure(figsize = (9, 9))
plt.plot(t, i(t), color = 'c', linewidth = 3, label = 'Current')
plt.plot(t, v(t), color = 'b', linewidth = 3, label = 'Voltage')

# Attributes
plt.title('LC Circuit with Piecewise Alternating Input Potential')
plt.xlabel('Time, t')
plt.ylabel('Current and Voltage')
plt.legend()

# Show the plot
plt.show()
```
