# Complex Numbers 

```python
import matplotlib.pyplot as plt
import cmath
import numpy as np
import sympy as sp
import pandas as pd
from IPython.display import display, Math
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Define some complex numbers
complex_numbers = [2 + 3j, -1 + 2j, -2 - 1j, 1 - 1j]

# Extract real and imaginary parts
real_parts = [z.real for z in complex_numbers]
imaginary_parts = [z.imag for z in complex_numbers]

# Plot the complex numbers
plt.scatter(real_parts, imaginary_parts, color='red')

# Add labels and title
plt.axhline(0, color='black', linewidth = 0.5)
plt.axvline(0, color='black', linewidth = 0.5)
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.title('Complex Numbers')

# Show plot
plt.grid(True, linewidth = 0.2)
plt.show()
```

## Complex Algebra

```python
# Defining two complex numbers
z1 = 2 + 3j
z2 = 3 - 5j

# Addition
print("Sum of the two numbers:", z1 + z2)

# Difference
print("Difference of the two numbers:", z1 - z2)

# Multiplication
print("Product:", z1*z2)

# Division
print("Ratio:", z1/z2)

# Conjugate
conjugate_z1 = z1.conjugate()
print("Conjugate of the first:", conjugate_z1)
conjugate_z2 = z2.conjugate()
print("Conjugate of the second:",conjugate_z2)

# Real and imaginary Parts
print("Real part of the first:", z1.real)
print("Real part of the second:", z2.real)
print("Imaginary part of the first:", z1.imag)
print("Imaginary part of the second:", z2.imag)

# Magnitude and Argument
print("Magnitude of the first:", abs(z1))
print("Argument of the first:", cmath.phase(z1))
print("Aegument of i:", cmath.phase(0 + 1j))

# Polar and Rectangular Form Conversion
print("Polar coordinates of i:", cmath.polar(0 + 1j)) #output's a tuple NOT a list
print("Polar coordinates of the first:", cmath.polar(z1))
modulus = 1.4142135623730951
phase = 0.7853981633974483
print("Third complex number in polar form:", (modulus, phase))
z3 = cmath.rect(modulus, phase)
print("Rectangular coordinates of the third:", z3)

# Exponentials, Logrithms and Powers
print("Exponential of the first:", cmath.exp(z1))
print("Logarithm of the first:", cmath.log(z1))
print("Second power of the first:", z1**2)

# Trignometric functions
print("Sine of the first:", cmath.sin(z1))
print("Cosine of the first:", cmath.sin(z1))
```

## Complex Function

```python
# Define the function f(z) = z^2
def f(z):
    return z**3 - 1

# Generate a grid of points in the complex plane
x = np.linspace(-2, 2, 400)
# print(type(x), "x:",x)
y = np.linspace(-2, 2, 400)
# print(type(y), "y:", y)
X, Y = np.meshgrid(x, y)
# print(type(X), "X:", X)
# print(type(Y), "Y:", Y)
Z = X + 1j * Y
# print(type(Z), "Z:", Z)

# Apply the function to the complex numbers
FZ = f(Z)
#print(type(FZ), "F(z):", FZ)

# Plot the magnitude of the function
plt.contourf(X, Y, np.abs(FZ), levels=20, cmap='viridis')
plt.colorbar(label='$|f(z)|$')
plt.xlabel('Real part of $z$')
plt.ylabel('Imaginary part of $z$')
plt.title('Magnitude of $f(z) = z^3 - 1 $')

# Show plot
plt.show()
```

```python
# Plotting in 3D

# Define the function f(z) = z^2
def f(z):
    return z**3 - 1

# Generate a grid of points in the complex plane
x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# Apply the function to the complex numbers
FZ = f(Z)

# Create a 3D plot
fig = plt.figure() # plt.figure() is used to create a new figure (or plotting area) in Matplotlib. 
# It acts as a container for all plot elements such as axes, labels, titles, etc.
ax = fig.add_subplot(111, projection='3d') # ax represents a set of axes in the figure fig.
# add_subplot() adds a subplot (or axes) to the figure. 
# The 111 argument specifies the layout of the subplot, and projection='3d' indicates that this subplot should be a 3D plot.
# 111: It means the subplot layout will be 1 row, 1 column, and youre creating the first (and only) subplot. 
# If you had more subplots, the first digit represents the number of rows, 
# the second represents the number of columns, and the third represents the position.
# projection='3d': This enables 3D plotting on this axes.

# Plot the magnitude of the function
ax.plot_surface(X, Y, np.abs(FZ), cmap='viridis')

# Add labels and title
ax.set_xlabel('Real part of $z$')
ax.set_ylabel('Imaginary part of $z$')
ax.set_zlabel('$|f(z)|$')
ax.set_title('3D Plot of $|f(z) = z^3 - 1|$')

# Show plot
plt.show()
```

### Finding Roots of Unity

```python
# Decalaration and Input
print("This code finds roots of unity in complex plane!")

z = complex((input("Enter the candidate:")))
n = int(input("Enter the radical of the candidate:"))

# Extracting polar coordinates
length = abs(z)
arg = cmath.phase(z)

# Root printing
w_list = [] # List of Roots
for k in range(0, n):
    w = (np.power(length, 1/n))*(np.exp(1j*(arg + 2*k*np.pi)/n))
    #print("Root", k+1, ":", w)
    
    #w_polar = cmath.polar(w)
    #print("Polar coordinates of root", k+1, ":", w_polar)
    
    w_list.append(w) # Appends the root to the list
    #print(w_list)
    
    plt.plot([0, w.real], [0, w.imag])
    
    #w = sp.exp(sp.I*2*sp.pi*k/n)
    #w_polar_sym = (sp.latex(w_polar[0]), sp.latex(w_polar[1]))
    #display(Math("%s %s" %(sp.latex(\zero), sp.latex(w))))
    
    
# Plotting roots in the Argand plane
plt.scatter([z.real for z in w_list], [z.imag for z in w_list], s =100, color='red')


#Define the radius and center of the circle
r = np.power(length, 1/n)       # Radius
center = (0, 0)  # Center of the circle (x, y)

# Generate angles from 0 to 2pi
theta = np.linspace(0, 2 * np.pi, 100)  # 100 points around the circle

# Parametric equations for the circle
x = center[0] + r * np.cos(theta)  # x-coordinates
y = center[1] + r * np.sin(theta)  # y-coordinates

# Create a scatter plot of the circle points
plt.plot(x, y, color='blue')
# plt.scatter(x, y, color='blue')



# Add labels and title
plt.axhline(0, color='black', linewidth = 0.5)
plt.axvline(0, color='black', linewidth = 0.5)
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.title('Complex Roots')
plt.gca().set_aspect('equal') # Set equal scaling for both axes


# Show plot
plt.grid(True, linewidth = 0.2)
plt.show()  

# Creating a dataframe using pandas
Unity_Roots = pd.DataFrame({'Real Part': [z.real for z in w_list], 'Imaginary Part': [z.imag for z in w_list]})
#print(Unity_Roots)
styled_Unity_Roots = Unity_Roots.style.set_table_styles([{'selector': 'th, td', 'props': [('border', '1px solid black')]}])
styled_Unity_Roots
```

```python
# Define a function to analyze the input function
def analyze_function(func_str):
    # Define the variable
    x = sp.symbols('x')
    
    # Convert the input string to a symbolic function
    func = sp.sympify(func_str)
    
    # Generate data for plotting
    x_values = np.linspace(-10, 10, 400)
    y_values = [func.subs(x, val) for val in x_values]  # Evaluate the function for each x

    # Plot the function
    plt.plot(x_values, y_values, label=f'f(x) = {func_str}')
    plt.axhline(0, color='black', linewidth=0.5, ls='--')
    plt.axvline(0, color='black', linewidth=0.5, ls='--')
    plt.title('Function Plot')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid()
    plt.show()

# Main function to run the program
def main():
    # Ask the user for a function
    user_input = input("Please enter a function of x (e.g., x**2 + 3*x + 2): ")
    analyze_function(user_input)

# Run the main function
if __name__ == "__main__":
    main()
```
